```python
import matplotlib.pyplot as plt

df['Hemoglobin'].hist()

plt.title("Hemoglobin Distribution")
plt.xlabel("Hemoglobin")
plt.ylabel("Frequency")

plt.show()
```

```python
print(df['Gender'].value_counts())
```

```python
#avg hemoglobin by gender
print(df.groupby('Gender')['Hemoglobin'].mean())
```

```python
#hemoglobin visualization acc to gender

gender_labels = ['Female', 'Male']

df.groupby('Gender')['Hemoglobin'].mean().plot(
    kind='bar',
    xticks=[0,1]
)

plt.xticks([0,1], gender_labels)

plt.title("Average Hemoglobin by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Hemoglobin")

plt.show()
```
